```python
import plasma_params as pp
import numpy as np
data = pytplot.data_quants['erg_mgf_l2_magt_8sec']
fce= np.zeros(data['time'].size)
fce1 = np.zeros(data['time'].size)
fce2 = np.zeros(data['time'].size)
for i in range(data['time'].size):
    fce2[i] = pp.Q*data[i]/pp.ME/1e9/2/np.pi/1e3*0.2
    fce1[i] = pp.Q*data[i]/pp.ME/1e9/2/np.pi/1e3*0.1
    fce[i] = pp.Q*data[i]/pp.ME/1e9/2/np.pi/1e3

pytplot.store_data('fce2', data={'x': data['time'], 'y': fce2})
pytplot.store_data('fce1', data={'x': data['time'], 'y': fce1})
pytplot.options('fce2', opt_dict={'ytitle':'ofa-B','ylog': 1, 'line_style': '-', 'Color': 'darkblue', 'thick': 1.2, 'alpha':1})
pytplot.options('fce1', opt_dict={'ylog': 1, 'line_style': '-', 'Color': 'forestgreen', 'thick': 1.2, 'alpha':1})
```

```python
pytplot.xlim('2017-03-27 21:30:00', '2017-03-27 22:00:00')
pytplot.tplot(['erg_mgf_l2_magt_8sec', 'fce1','fce2'])
```
