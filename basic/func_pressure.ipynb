```python
import math
#energy = [184.2, 184.2, 148.9, 120.4,  97.3,  78.7,  63.7,  51.5,  41.7,
#        33.7,  27.3,  22.1,  17.9,  14.5,  11.8,   9.6]
energy = [109.6, 109.6, 87.8, 70.3, 56.3, 45.1, 36.2, 29.0, 23.3, 18.6, 15.0, 12.0, 9.7, 7.8, 6.3, 5.1]
n = len(energy)

dlt_enrg = []
for i in range(n):
    if i == 0:
        delt_enrg_i = math.sqrt(energy[i] * energy[i]) - math.sqrt(energy[i] * energy[i+1])
    elif i == n-1:
        delt_enrg_i = math.sqrt(energy[i-2] * energy[i-2]) - math.sqrt(energy[i-1] * energy[i-2])
    else:
        delt_enrg_i = math.sqrt(energy[i-1] * energy[i]) - math.sqrt(energy[i] * energy[i+1])
    dlt_enrg.append(delt_enrg_i)


```

```python


#dlt_enrg[0]=20
#dlt_enrg[1]=20
dlt_enrg    
```

```python
import plasma_params as pp
import numpy as np

mepifp_pressure = np.zeros(len(mepifp_add['time']))
mepifhe2_pressure = np.zeros(len(mepifhe2_add['time']))
mepihe_pressure = np.zeros(len(mepihe_add['time']))
mepifopp_pressure = np.zeros(len(mepifopp_add['time']))
mepifo_pressure = np.zeros(len(mepifo_add['time']))
mepifo2p_pressure = np.zeros(len(mepifo2p_add['time']))

mp = 1.6725*1e-27
mhe = 1.67262192*1e-27
mhe2 = 2*mp
mo = 2.6566*1e-26
mo2 = 2*mo

for i in range(len(mepifp_add['time'])):
    for j in range(len(energy)):
        mepifp_pressure[i] += np.sqrt(2 * mp * energy[j]) * mepifp[i][j] * dlt_enrg[j]
        mepifhe2_pressure[i] += np.sqrt(2 * mhe2 * energy[j]) * mepifhe2[i][j] * dlt_enrg[j]
        mepihe_pressure[i] += np.sqrt(2 * mhe * energy[j]) * mepihe[i][j] * dlt_enrg[j]
        mepifopp_pressure[i] += np.sqrt(2 * mo * energy[j]) * mepifopp[i][j] * dlt_enrg[j]
        mepifo_pressure[i] += np.sqrt(2 * mo * energy[j]) * mepifo[i][j] * dlt_enrg[j]
        mepifo2p_pressure[i] += np.sqrt(2 * mo2 * energy[j]) * mepifo2p[i][j] * dlt_enrg[j]


mepifp_p = (4/3) * np.pi*mepifp_pressure*1e-4
mepifhe2_p = (4/3) * np.pi*mepifhe2_pressure*1e-4
mepihe_p = (4/3) * np.pi*mepihe_pressure*1e-4
mepifopp_p = (4/3) * np.pi*mepifopp_pressure*1e-4
mepifo_p = (4/3) * np.pi*mepifo_pressure*1e-4

```

```python
plasma_pressure = (mepifp_p + mepifhe2_p + mepihe_p + mepifopp_p + mepifo_p) * 4 * np.pi * 10**(-1.5)
```

```python
pytplot.store_data('plasma_pressure', data={'x':mepifp_add['time'] , 'y': plasma_pressure})
```
