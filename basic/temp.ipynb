```python
import pyspedas
import pytplot
```

```python
tr = ['2017-03-29 01:00:00','2017-03-29 01:30:00']

from pyspedas.erg import mgf
mgf(trange=tr)
```

```python
pytplot.split_vec('erg_mgf_l2_mag_8sec_gse')
```

```python
import numpy as np
x = pytplot.data_quants['erg_mgf_l2_mag_8sec_gse_x']
y = pytplot.data_quants['erg_mgf_l2_mag_8sec_gse_y']
z = pytplot.data_quants['erg_mgf_l2_mag_8sec_gse_z']
total = np.sqrt(x**2 + y**2 + z**2)
pytplot.store_data('erg_mgf_l2_mag_8sec_gse_total', data={'x': pytplot.data_quants['erg_mgf_l2_mag_8sec_gse_x']['time'], 'y': total})
```

```python
pytplot.tplot(['erg_mgf_l2_mag_8sec_gse_total', 'erg_mgf_l2_magt_8sec', 'erg_mgf_l2_mag_8sec_gse_z'])
```
